```python
#必要なライブラリを呼ぶ
import pandas as pd
import numpy as np
from sklearn import linear_model

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df1 = pd.read_csv('/content/drive/My Drive/t1_t2,t3.csv')
df2 = pd.read_csv('/content/drive/My Drive/t3_t1,t2.csv')
```

```python
sns.pairplot(df1[['t2', 't3', 'p1','p2','t1']])
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df2[['t1', 't2', 'p1','p2','t3']])
plt.tight_layout()
plt.show()
```

```python
#t1からt2を生成する関数の作成
X_df1a = df1[['t1']]
y_df1a = df1['t2']
reg = linear_model.LinearRegression()
reg.fit(X_df1a, y_df1a)
print("重み:",reg.coef_) #重み
print("切片:",reg.intercept_) #切片
print("スコア:",reg.score(X_df1a, y_df1a))
```

```python
#t1からt3を生成する関数の作成
X_df1b = df1[['t1']]
y_df1b = df1['t3']
reg = linear_model.LinearRegression()
reg.fit(X_df1b, y_df1b)
print("重み:",reg.coef_) #重み
print("切片:",reg.intercept_) #切片
print("スコア:",reg.score(X_df1b, y_df1b))
```
